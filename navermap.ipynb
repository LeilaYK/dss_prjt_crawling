```python
# 역이름
station = "부평구청역"
url = "https://map.naver.com/v5/api/search?caller=pcweb&query={} 맛집&type=all&searchCoord=126.71639442443849;37.508432213060516&page=1&displayCount=20&isPlaceRecommendationReplace=true&lang=ko".format(station)
print(url)
response = requests.get(url)
response.json()
```

```python
r1 = response.json()["result"]["place"]["list"]
r1
```

```python
datas = response.json()["result"]["place"]["list"]
datas[0]
```

```python
ranks = [data["rank"] for data in datas]
names = [data["name"] for data in datas]
tels = [data["tel"] for data in datas]
addrs = [data["address"] for data in datas]
categories = [data["category"] for data in datas]
menus = [data["menuInfo"] for data in datas]
reviews = [data["microReview"] for data in datas]
```

```python
columns = ["ranks", "names", "tels", "addrs", "categories", "menus", "reviews"]
info = pd.DataFrame(columns=columns)
info["ranks"] = ranks
info["names"] = names
info["tels"] = tels
info["addrs"] = addrs
info["categories"] = categories
info["menus"] = menus
info["reviews"] = reviews
```
